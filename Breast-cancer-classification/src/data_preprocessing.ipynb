```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
```

```python
# read raw data
PATH = '../data/'
df = pd.read_csv(PATH + 'raw_data.csv')
df.head()
```

# Preprocessing

```python
df.dtypes
```

```python
# check for null values
df.isnull().sum()
```

```python
# check for duplicated rows
len(df[df.duplicated()])
```

```python
df['diagnosis'] = df['diagnosis'].replace(['B', 'M'], [0, 1])
df.head()
```

```python
df.describe()
```

# Analyze

```python
# plot correlation matrix
correlation_matrix = df.corr()
plt.figure(figsize=(12, 12))
sns.heatmap(correlation_matrix, center=0)
```

```python
correlation_matrix
```

```python
# drop rows which have no dirrect correlation with the diagnosis
lower_bound = -0.1
upper_bound = +0.1
for col in correlation_matrix.columns:
    if lower_bound <= correlation_matrix.loc['diagnosis', col] <= upper_bound:
        df.drop(col, axis=1, inplace=True)
```

```python
df.shape
```

```python
# save cleaned data file
df.to_csv(PATH + 'clean_data.csv', index_label=False)
```
